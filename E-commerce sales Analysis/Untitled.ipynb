```python
Find = df1["Product"].value_counts()
plt.bar(Find.index , Find.values , color = "pink" , edgecolor = "Black" , label = "sales of products")
plt.xlabel("Product")
plt.ylabel("sales")
plt.title("Analysis the sales of products")
for i in range(len(Find)):
    plt.text(Find.index[i], Find.values[i] + 1 , str(Find.values[i]) , ha = 'center')
plt.show()
```
